
The Dataset is converted into two sets based on region from 122th index,, 
we can make a new column based on the region

1 : "Bejaia Region Dataset"
2 : "Sidi-Bel Abbes Region Dataset"

Add new column with region

```python
dataset.loc[:122 , "Region"] = 0
dataset.loc[122: , "Region"] = 1
```

```python
df = dataset.copy()
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#### Dropping a row which isn't useful

df = df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
#### Fix the spaces in the dataset

df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
#### Changing the required column datatypes to int , float as necessary

df.columns
```

```python
df[['day' , 'month' , 'year' , 'Temperature' , 'RH' , 'Ws']] = df[['day' , 'month' , 'year' , 'Temperature' , 'RH' , 'Ws']].astype(int)
```

```python
df.info()
```

```python
df.columns
```

```python
objects = [features for features in df.columns if df[features].dtype == 'O']
```

```python
objects
```

```python
for i in objects:
    if i != 'Classes':
        df[i] = df[i].astype(float)
```

```python
df['Classes'] = np.where(df['Classes'].str.contains('not fire') , 'not fire' , 'fire')
```

```python
#### Monthly Fire Analysis

dftemp = df.loc[df['Region'] == 1]
plt.subplots(figsize=(13 , 6))
sns.set_style('whitegrid')
sns.countplot(x='month' , hue='Classes' , data=df)
plt.ylabel('Number of Fires' , weight='bold')
plt.xlabel('Months' , weight='bold')
plt.title('Monthly Fire Analysis in Regions')
```

```python
#### Monthly Not Fire Analysis

dftemp = df.loc[df['Region'] == 0]
plt.subplots(figsize=(13 , 6))
sns.set_style('whitegrid')
sns.countplot(x='month' , hue='Classes' , data=df)
plt.ylabel('Number of Fires' , weight='bold')
plt.xlabel('Months' , weight='bold')
plt.title('Monthly Not Fire Analysis in Regions')
```

**Its Observed that august and september had the most no of forest fires for both regions. And from the above plot of months we can understand few things**

Most of the fires happened in august and very high fires happened in only 3 months - June , July , August

Less fire was on september
